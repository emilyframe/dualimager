```python
# HEU Scan Using CCI2
%matplotlib qt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
import re

E = 186
sigma = 2
line_data = np.zeros((23, 3))
i = 0
for file in glob.glob('/Users/eframe/dmi/data/heuscan_CCI2/*.h5'):
    section = re.search(r'out(.*?)_', file).group(1) # starting point in (mm)
    print(file, section)
    f = tables.open_file( file, 'r' )
    time = f.root.EventData.read()['timestamp']
    duration = ( time[-1] * 10 ** -8 - time[0] * 10 ** -8 ) # seconds
    l11 = f.root.Interactions.Single.read()

    # Getting Cone Data
    emask = ( l11['energy'] >= E - sigma ) & ( l11['energy'] <= E + sigma ) 
    emaskR = ( l11['energy'] < E - sigma ) & ( l11['energy'] >= E - 2 * sigma ) 
    emaskL = ( l11['energy'] < E + sigma ) & ( l11['energy'] >= E + 2 * sigma ) 
    events= l11[emask]
    bevents = len(l11[emaskR]) + len(l11[emaskL])
    cps = ( len( events ) - bevents ) / duration
    err = np.sqrt( len( events ) - bevents ) / duration
    line_data[i][0] = section
    line_data[i][1] = cps
    line_data[i][2] = err
    i = i + 1
line_data = line_data[line_data[:,0].argsort()]
 
    
font = FontProperties()
font.set_name('Times New Roman')
fig, ax = plt.subplots()
ax.minorticks_on()
ax.tick_params(which='major', length=10, width=2, direction='in', labelsize=50)
ax.tick_params(which='minor', length=5, width=2, direction='in')
ax.set_xlabel('Lateral Distance [mm]', fontsize=50, fontproperties=font)
ax.set_ylabel('Counts per Second', fontsize=50, fontproperties=font)
ax.errorbar( line_data[:,0] + 2.5 - 10, line_data[:,1], yerr=line_data[:,2], marker='s', capsize=10, elinewidth=3,  mfc='k', color='k', label='Error')
ax.legend(prop={"family":"Times New Roman", 'size': 50}, loc='upper left', frameon=False )
ax.tick_params(labelsize=50)
for l in ax.xaxis.get_ticklabels():
    l.set_family("Times New Roman")
    l.set_size(50)
for l in ax.yaxis.get_ticklabels():
    l.set_family("Times New Roman")
    l.set_size(50)
# ax.grid(which='both')
plt.show()
```
